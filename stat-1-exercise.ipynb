```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("songs_normalize.csv")
```

```python
df.head() 
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='year',y='popularity',data=df)
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='year',y='popularity',data=df)
```
